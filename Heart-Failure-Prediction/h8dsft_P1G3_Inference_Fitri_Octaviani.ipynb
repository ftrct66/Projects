# Import Library

```python
# import library
import pandas as pd
import numpy  as np
import joblib
```

# Load Model

```python
# load model
with open('rf_tuning.pkl','rb') as file_1:
    rf_tuning = joblib.load(file_1)

with open('gbc_tuning.pkl','rb') as file_2:
    gbc_tuning = joblib.load(file_2)
```

# Data Loading

```python
# fitur
fitur = ['age', 'anaemia', 'creatinine_phosphokinase', 	'high_blood_pressure',	'serum_creatinine']

age = [25, 50]
anaemia = [1, 0]
creatinine_phosphokinase = [143, 200]
high_blood_pressure = [0, 1]
serum_creatinine = [2.50, 1.90]

data1 = {
    'age':age,
    'anaemia':anaemia,
    'creatinine_phosphokinase':creatinine_phosphokinase,
    'high_blood_pressure':high_blood_pressure,
    'serum_creatinine':serum_creatinine
}
df_infer = pd.DataFrame(data=data1,columns=fitur)
```

```python
df_infer.head()
```

# Model Inference

```python
# predict result random forest model
pred_rf = rf_tuning.predict(df_infer)

# predict result gradientboost model
pred_gbc = gbc_tuning.predict(df_infer)
```

```python
# show data infer + prediction
df_infer['pred_target_rf'] = pred_rf
df_infer['pred_target_gbc'] = pred_gbc
df_infer
```
