## **SOAL 1**

> Berapa jumlah transaksi yang berstatus Complete tiap bulan selama Q1 sampai Q3 di tahun 2022?

```python
# menghitung jumlah transaksi
Jumlah_transaksi = client.query('''
SELECT
    # extract tahun, bulan, dan quarter dari kolom created_at
    EXTRACT (year FROM created_at) as Tahun,
    EXTRACT (month FROM created_at) as Bulan,
    EXTRACT (QUARTER FROM created_at) as Quarter,
    # menghitung jumlah transaksi berdasarkan status
    COUNT (status) as Jumlah_Transaksi
FROM bigquery-public-data.thelook_ecommerce.order_items
WHERE
    # Kondisi dimana status yang diambil adalah Complete
    status = 'Complete'
    # Tahun 2022, Bulan 1 sampai 9, Quarter kurang dari 4
    AND EXTRACT (year FROM created_at) = 2022
    AND EXTRACT (month FROM created_at) <= 9
    AND EXTRACT(QUARTER FROM created_at) < 4
# mengelompokkan baris
GROUP BY Tahun, Bulan, Quarter
ORDER BY Bulan

''').to_dataframe()
Jumlah_transaksi
```

**INSIGHT**

Data yang memiliki jumlah transaksi yang berstatus Complete tersedia pada tabel order items. Untuk menghitung status Complete tersebut perlu menggunakan aggregasi COUNT, dan quartil yang digunakan adalah kolom created_at menggunakan fungsi EXTRACT untuk memisahkan Q1 sampai Q3 dengan menggunakan data bigquery-public-data.thelook_ecommerce.order_items.

- Jumlah total pengguna yang berstatus complete selama 3 quartil adalah 10.696 users.

- Jumlah total pengguna terendah tercatat pada bulan ke-2 sebesar 902 users, sedangkan jumlah pengguna tertinggi terjadi pada bulan ke-9 sebanyak 1483.

- Jumlah transaksi yang berstatus Complete semakin meningkat dari tiap bulan atau quartalnya, hal ini merupakan indikator positif untuk bisnis dan dapat lebih ditingkatkan lagi agar mencapai targat yang diinginkan.

## **SOAL 2**

> Berikan informasi total penjualan (dalam USD) tiap bulan selama Q1 sampai Q3 di tahun 2022! (Hanya yang transaksi berstatus Complete).







```python
# Menghitung total penjualan tiap bulan
total_penjualan = client.query('''
SELECT
  # extract tahun, bulan, dan quarter dari kolom shipped_at
  EXTRACT(YEAR FROM shipped_at) as Tahun,
  EXTRACT(MONTH FROM shipped_at) as Bulan,
  EXTRACT (QUARTER FROM shipped_at) as Quarter,
  # melakukan perhitungan pada sekumpulan nilai dari product_retail_price
  SUM(product_retail_price) AS Total_penjualan
FROM
  `bigquery-public-data.thelook_ecommerce.order_items` AS items
JOIN
  `bigquery-public-data.thelook_ecommerce.inventory_items` AS inventory
# spesifikasi kondisi join dari dua dataset
ON
  items.product_id = inventory.product_id
WHERE
  status = 'Complete'
  AND EXTRACT(YEAR FROM shipped_at) = 2022
  AND EXTRACT(MONTH FROM shipped_at) BETWEEN 1 AND 9
  AND EXTRACT(QUARTER FROM shipped_at) < 4
GROUP BY Tahun, Bulan, Quarter
ORDER BY Bulan

''').to_dataframe()
total_penjualan
```

**INSIGHT**

Untuk menghitung total penjualan tiap bulan diperlukan 2 dataset dengan menggunakan fungsi JOIN, order_items dan inventory_items. Setelah itu menggunakan aggregasi SUM untuk menghitung total penjualan tiap bulan pada product_retail_price dengan status complete pada Q1 sampai Q3 tahun 2022.

- Total pendapatan selama 3 quartil adalah 12849626 USD.

- Pendapatan terendah tercatat pada bulan ke-2 sebesar 1.117165 USD, sementara pendapatan tertinggi terjadi pada bulan ke-9 sebanyak 1.775233 USD.

- Dikarenakan pada bulan ke-dua dan ke-empat sempat mengalami penurunan jumlah transaksi yang berstatus Complete, sehingga total penjualan dalam USD yang diperoleh dari bulan ke-dua dan ke-empat pun mengalami penurunan dalam total penjualan (USD). Akan tetapi untuk bulan selanjutnya mengalami peningkatan dalam total penjualan (USD).

## **SOAL 3**

> Berapa user yang melakukan transaksi berstatus Complete di tiap bulan dari Q1 sampai Q3 2022?

```python
# menghitung user yang melakukan transaksi di tiap bulan
user_transaksi = client.query('''
SELECT
    EXTRACT (year FROM items.created_at) as Tahun,
    EXTRACT (month FROM items.created_at) as Bulan,
    EXTRACT (quarter FROM items.created_at) as Quarter,
    # hanya values yang berbeda agar data users yang sama tidak berulang kali
    COUNT (DISTINCT user_id) as Users
FROM `bigquery-public-data.thelook_ecommerce.users` as users
JOIN `bigquery-public-data.thelook_ecommerce.order_items` as items
ON users.id = items.user_id
WHERE
    status = 'Complete'
    AND EXTRACT(YEAR FROM items.created_at) = 2022
    AND EXTRACT(MONTH FROM items.created_at) BETWEEN 1 AND 9
    AND EXTRACT(QUARTER FROM items.created_at) < 4
GROUP BY Tahun, Bulan, Quarter
ORDER BY Bulan

''').to_dataframe()
user_transaksi
```

**INSIGHT**

Dalam menghitung user yang melakukan transaksi di tiap bulan, dataset yang digunakan adalah users dan order_items menggunakan fungsi JOIN, hanya saja pada aggregasi COUNT ini menggunakan DISTINCT agar user_id yang sama tidak dihitung secara berkala.

- User yang melakukan transaksi terendah tercatat pada bulan ke-2 sebesar 620 users, sementara users yang melakukan transaksi terbanyak tercatat pada bulan ke-9 sebanyak 1067.

- Dalam hal inipun memengaruhi total penjualan (USD) yang diperoleh, semakin banyak user yang bertransaksi maka semakin banyak pula pengahasilan total penjualan (USD). Dalam segi positif berarti promosi yang dilakukan oleh e-commerce ini dapat menarik perhatian users.

## **SOAL 4**

> Kategori produk apa saja yang paling banyak dibeli (status transaksi: Complete) di tiap bulannya selama Q1 sampai Q3 tahun 2022?

```python
# Melihat kategori produk yang paling banyak di beli
kategori_produk = client.query('''
SELECT Tahun, Bulan, Quarter, category AS Kategori, total AS Total_dibeli
FROM (
    SELECT
        Tahun, Bulan, Quarter, category, total,
        ROW_NUMBER() OVER (PARTITION BY Bulan ORDER BY total DESC) AS Partisi
    FROM (
        SELECT
            category,
            EXTRACT(Year FROM created_at) AS Tahun,
            EXTRACT(MONTH FROM created_at) as Bulan,
            EXTRACT(Quarter FROM created_at) AS Quarter,
            COUNT(products.id) as total
        FROM `bigquery-public-data.thelook_ecommerce.products` as products
        JOIN `bigquery-public-data.thelook_ecommerce.order_items` as items
        ON products.id = items.product_id
        WHERE
            status = 'Complete' AND
            created_at BETWEEN '2022-01-01' AND '2022-09-30'
        GROUP BY
            Tahun, Bulan, Quarter, category
        ) AS subs
) AS subs2
WHERE Partisi = 1
ORDER BY Bulan

''').to_dataframe()
kategori_produk
```

**INSIGHT**

Fungsi yang digunakan untuk melihat kategori produk yang paling banyak di beli adalah window function dengan OVER partition by. Partisi ini di convert berdasarkan bulan dan kategori produk.

- Dalam kategori produk memiliki macam-macam ketersediaan, diantaranya Tops & Tees, Maternity, Sweater, Jeans, Outerwear & Coats, dan lain sebagainya. Dalam 3 quartil tersebut, produk yang paling sering menjadi tren atau paling sering di beli adalah Intimates selama 5 bulan.

- Agar kategori produk lain mengalami kenaikan tren, maka perlu melakukan promosi atau pemasaran lebih, melakukan inovasi terhadap kategori yang kurang tren agar menarik perhatian users.

## **SOAL 5**

> Kategori produk apa saja yang paling banyak mendapatkan pendapatan (status transaksi: Complete) di tiap bulannya selama Q1 sampai Q3 tahun 2022?

```python
# Melihat dan menghitung pendapatan paling banyak dari kategori produk
pendapatan_produk = client.query('''
SELECT
    Year,
    Month,
    Quarter,
    category AS Kategori,
    total_sales AS Total_Pendapatan
FROM (SELECT *,
            DENSE_RANK() OVER (PARTITION BY Month ORDER BY total_sales DESC) as rank
      FROM (SELECT
              EXTRACT(Year FROM created_at) AS Year,
              EXTRACT(MONTH FROM created_at) AS Month,
              EXTRACT(Quarter FROM created_at) AS Quarter,
              products.category,
              SUM(sale_price) AS total_sales,
            FROM `bigquery-public-data.thelook_ecommerce.order_items` AS items
            JOIN `bigquery-public-data.thelook_ecommerce.products` AS products
            ON items.product_id = products.id
            WHERE status='Complete'
                  AND EXTRACT(YEAR FROM created_at) = 2022
                  AND EXTRACT(MONTH FROM created_at) BETWEEN 1 AND 9
                  AND EXTRACT(QUARTER FROM created_at) < 4
            GROUP BY Year, Month, Quarter, products.category)
                     )
WHERE rank=1
ORDER BY Month''') \
.to_dataframe()
pendapatan_produk
```

**INSIGHT**

Dalam kategori produk yang paling banyak dibeli rata-rata tiap bulan memang Intimates, namun untuk produk yang paling banyak menghasilkan pendapatan di tiap bulannya bukan Intimates, melainkan Outerwear & Coats pada bulan ke- 1, 2, 3, 4, 6, 8, 9, dan Jeans pada bulan ke 5 dan 7. Hal ini bisa saja terjadi karena harga masing-masing kategori berbeda.Dalam kurun waktu 9 bulan tersebut, total pendapatan tertinggi adalah bulan 6 dengan kategori Outerwear & Coats	yang menghasilkan total pendapatan sebesar 10529.359953.
